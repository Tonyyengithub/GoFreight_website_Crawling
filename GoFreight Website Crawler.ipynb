```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://www.gofreight.co/blog'
web = requests.get(URL)
soup = BeautifulSoup(web.text, "html.parser")
```

```python
soup
```

```python
#content = soup.find_all('a', class_ = "title")
contents = soup.find_all('a', {'class': 'title'})
A = []
i = list(range(0,27)) # 26篇文章
# print(i)
for index in i:
    c = contents[index].string.strip()
    # .string 只取 page 的 index
    # .strip() 把多餘空格都去除
    A.append(c)
print(A)
```

```python
# 取 a 裡面的網址
web = []
for i in contents:
    if "https" in str(i['href']):
            web.append(str(i['href']))
print(web)
```

```python
df = {'Page title':[], 'URL Link':[]}

for url in web:
    url = url.split("\n")[0]
    df['URL Link'].append(url)

for index in A:
    print(index)
    index = index.split("\n")[0]
    df['Page title'].append(index)

# df = pd.DataFrame.from_dict(df, orient = 'index') # 橫的
df = pd.DataFrame.from_dict(df)
df.to_csv("file.xlsx", index = False)
```
